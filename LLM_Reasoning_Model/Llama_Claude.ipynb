```python
import torch
from transformers import pipeline

# 1. Khai báo ID của model
model_id = "DavidAU/Llama3.3-8B-Instruct-Thinking-Claude-4.5-Opus-High-Reasoning"

print(f"Đang tải model: {model_id}...")

# 2. Khởi tạo pipeline
# device_map="auto" sẽ tự động dùng GPU nếu có
# torch_dtype=torch.bfloat16 giúp giảm VRAM và chạy nhanh hơn trên GPU đời mới
pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

# 3. Tạo nội dung chat
# Model này có khả năng "Thinking" (suy luận), nên hãy hỏi những câu cần tư duy
messages = [
    {
        "role": "system",
        "content": "You are a helpful AI assistant capable of deep reasoning."
    },
    {
        "role": "user",
        "content": "Hãy giải thích ngắn gọn: Tại sao bầu trời lại có màu xanh? Trả lời bằng tiếng Việt."
    },
]

# 4. Chạy model (Inference)
print("Model đang suy nghĩ...")
outputs = pipe(
    messages,
    max_new_tokens=512,  # Độ dài tối đa câu trả lời
    do_sample=True,      # Cho phép ngẫu nhiên để văn bản tự nhiên hơn
    temperature=0.7,     # Độ sáng tạo (0.7 là mức cân bằng tốt)
    top_p=0.9,
)

# 5. In kết quả
# Lấy phần trả lời của Assistant ra khỏi chuỗi hội thoại
response = outputs[0]["generated_text"][-1]["content"]
print("-" * 50)
print("🤖 CÂU TRẢ LỜI CỦA MODEL:\n")
print(response)
print("-" * 50)
```

```python
import torch
from transformers import pipeline

# 1. Model ID
model_id = "DavidAU/Llama3.3-8B-Instruct-Thinking-Claude-4.5-Opus-High-Reasoning"

print(f"Loading model: {model_id}...")

# 2. Initialize Pipeline
# device_map="auto" automatically uses GPU
# torch_dtype=torch.bfloat16 reduces VRAM usage and speeds up inference on modern GPUs
pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

# 3. Create Chat Content
# Switching prompt to English for faster token generation
messages = [
    {
        "role": "system",
        "content": "You are a helpful AI assistant capable of deep reasoning."
    },
    {
        "role": "user",
        "content": "Explain briefly: Why is the sky blue?"
    },
]

# 4. Run Inference
print("Model is thinking...")
outputs = pipe(
    messages,
    max_new_tokens=512,  # Maximum length of generated text
    do_sample=True,      # Enable sampling for more natural text
    temperature=0.7,     # Creativity level
    top_p=0.9,
)

# 5. Print Result
response = outputs[0]["generated_text"][-1]["content"]
print("-" * 50)
print("🤖 MODEL RESPONSE:\n")
print(response)
print("-" * 50)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig

# 1. Cấu hình Model ID
model_id = "DavidAU/Llama3.3-8B-Instruct-Thinking-Claude-4.5-Opus-High-Reasoning"

print("Đang khởi động lại và tải model...")

# 2. Cấu hình Quantization (Nén 4-bit)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
)

# 3. Tải Tokenizer trước
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 4. Tải Model thủ công (Để kiểm soát lỗi VRAM tốt hơn)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto", # Tự động chia GPU
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True
)

# 5. Đưa Model vào Pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

# 6. Tạo Prompt (Kèm System Prompt kích hoạt tư duy)
messages = [
    {
        "role": "system",
        "content": "You are an advanced AI assistant. You must think step-by-step inside <thinking> tags before answering."
    },
    {
        "role": "user",
        "content": "Explain briefly: How does a neural network learn?"
    },
]

# 7. Chạy Inference
print("Model is thinking...")
outputs = pipe(
    messages,
    max_new_tokens=512,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)

print("-" * 50)
print(outputs[0]["generated_text"][-1]["content"])
print("-" * 50)
```
